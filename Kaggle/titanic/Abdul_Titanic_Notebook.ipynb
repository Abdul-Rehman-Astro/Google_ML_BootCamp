```python
import pandas as pd
```

```python
training = pd.read_csv("train.csv")
testing = pd.read_csv("test.csv")

testID = testing["PassengerId"]

training.head(10)
```

```python
def clean(data):
    data = data.drop(["Name","Ticket","Cabin","PassengerId"], axis= 1)

    cols = ["SibSp", "Parch" , "Fare","Age"]

    for col in cols:
        data[col].fillna(data[col].median(),inplace=True)

    data.Embarked.fillna("U",inplace=True)
    
    return data
```

```python
training = clean(training)
testing = clean(testing)
```

```python
training.head(5)
```

Now lets incode male to 0 and female to 1 

```python
from sklearn import preprocessing

# Here we introduce lable encoder as le
le =preprocessing.LabelEncoder()

columns = ["Sex", "Embarked"]

for col in columns:
    training[col] = le.fit_transform(training[col])
    testing[col]  = le.transform(testing[col])
    print( le.classes_)

training.head(5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

Y = training["Survived"]
X= training.drop("Survived",axis =1)

X_train, X_validation,Y_train, Y_validation = train_test_split(X,Y,test_size = 0.2, random_state = 42)
```

```python
clf = LogisticRegression(random_state= 0 ,max_iter= 1000).fit(X_train, Y_train)
```

```python
predictions = clf.predict(X_validation)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_validation,predictions)
```

```python
submit_preds = clf.predict(testing)
```

```python
df = pd.DataFrame({"PassengerID": testID.values,
                    "Survived": submit_preds
                    })
```

```python
df.to_csv("My_submission.csv",index = False )
```
